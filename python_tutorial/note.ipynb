```python
total = 0
jjak_hqp = 0
hole_hap = 0

for i in range(1, 101):
    total += i
    if i % 2 == 0:
        jjak_hqp  += i
    else:
        hole_hap += i

print(f"1~100 전체 합계: {total}")
print(f"1~100 짝수 합계: {jjak_hqp}")
print(f"1~100 홀수 합계: {hole_hap}")
```
